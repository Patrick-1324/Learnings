# Matplotlib
### Goal
- Learn to visualize data charts

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4])
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 5, 7, 10, 4, 8]
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Title")
ax.set_xlabel("X lable")
ax.set_ylabel("Y lable")
```

```python
data = [1,1,2,3,4,1,2,4,5,3,4,5,4,6,5,7,7,7,8]

fig, ax = plt.subplots()
ax.hist(data, bins = 8)
```
